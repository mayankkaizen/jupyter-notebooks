```python
from nltk.corpus import gutenberg
for f in gutenberg.fileids():
    num_chars = len(gutenberg.raw(f))
    num_words = len(gutenberg.words(f))
    num_sents = len(gutenberg.sents(f))
    num_vocabs = len(set(w.lower() for w in gutenberg.words(f)))
    print(round(num_chars/num_words), round(num_words/num_sents), round(num_words/num_vocabs), f)
    
#avg word length, avg sentence length, and lexical diversity score per text    
```

Above program shows some stats. Observe that avg word length seems to be general property of English language (here space is also included in numbers of characters). But avg sentence length and lexical diversity score seem specific to respective authors.

### Lexical Resources

Refer NLTK book, second chapter, same section

```python
def unusual_words(text):
    text_vocab = set(w.lower() for w in text if w.isalpha())
    eng_vocab = set(w.lower() for w in nltk.corpus.words.words())
    unusual = text_vocab - eng_vocab
    return sorted(unusual)

unusual_words(nltk.corpus.gutenberg.words('austen-sense.txt'))[:25]
```

There is also a corpus of stopwords, that is, high-frequency words like 'the', 'to' and also that we sometimes want to filter out of a document before further processing. Stopwords usually have little lexical content, and their presence in a text fails to distinguish it from other texts.

```python
from nltk.corpus import stopwords
stopwords.words('english')[:25]
```

Let's define a function to compute what fraction of words in a text are *not* in the stopwords list


```python
def content_fraction(text):
    stopwords = nltk.corpus.stopwords.words('english')
    content = [w for w in text if w.lower() not in stopwords]
    return len(content)/len(text)

content_fraction(nltk.corpus.reuters.words())
```

#### A Puzzle

Consider the following string of alphabets : 'EGIVRVONL' . Find out all the words which fulfil following conditions - 

  - Word length should be more than or equal to 6. 
  - Each letter may be used once per word
  - Each word must contain 'R'
  - There must be a 9 letter word
  - No plurals (ending with 's'), no foreign word, no proper nouns are allowed
  

```python
puzzle_letters = nltk.FreqDist('egivrvonl')
puzzle_letters
```

```python
obligatory = 'r'
wordlist = nltk.corpus.words.words()
a = [w for w in wordlist if len(w)>= 6 and obligatory in w and nltk.FreqDist(w) <= puzzle_letters]
```

NLTK includes the CMU Pronouncing Dictionary for US English which was designed for use by speech synthezisers.

In each entry, first is the word. This word is followed by a list which essentially contains information about phonetic codes known as phones. The symbol in CMU dictionarary are from [Arpabet](http://en.wikipedia.org/wiki/Arpabet).

In above example, notice the word 'fire'. It has 2 entries for its pronunciation. 

```python
for word, pron in entries:
    if len(pron) == 3:
        ph1, ph2, ph3 = pron
        if ph1 == 'P' and ph3 == 'T':
            print(word, ph2, end =' ')
```

The above program scans the lexicon looking for entries whose pronunciation consists of three phones.
